```python
# Keyword mode, defaults to 'add' if we do keyword = value.
# Keywords can be invoked in any order
def some_func(param1, param2, mode="add"):
    if mode == "add":
        return param1 + param2
    elif mode == "sub": # Else-if
        return param1 - param2
    #return None
    # Functions implicitly return None at the end code execution gets there.
    # You can use 'else' in addition to the above to have a default fallback case.
```

```python
print(some_func(10, 15, mode="add"))
```
