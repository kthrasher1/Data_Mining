# Seaborn Visualisation

### Task 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('train.csv')

embarked = df['Embarked']

embarked.hist(bins=5)
```

### Task 2

```python
sns.catplot(x='Age', y='Pclass', hue="Survived", data=df, palette='Set3')
```

```python
sns.catplot(x = 'Pclass', kind='count', data=df, palette='Set3')
```

### Task 3

```python
sns.boxplot(x=df['Pclass'], palette='Set3')
```

```python
ax = sns.boxplot(x='Fare', y='Embarked', data=df, palette='Set3')
ax.set_xscale('log')
```

```python
ax = sns.boxplot(x='Survived', y='Age', data=df, palette='Set3')
```

```python
ax = sns.boxplot(x='Sex', y='Age', data=df, palette='Set3')
```
